```python
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
```

```python
url = 'https://bcs-express.ru/investidei?status=executed'
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(5)  # Ожидание прогрузки страницы до 5 секунд
```

```python
while True:
    data = {'strategy_title': [], 'href': []}
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')


    try:
        
        button_element = driver.find_element(By.CSS_SELECTOR, "[data-id='button-more']")
        
        # Клик кнопки
        button_element.click()

        # Даем время для загрузки новых элементов
        time.sleep(2)
    
    except NoSuchElementException:
        # Если кнопка больше не найдена
        print("Кнопка 'Показать еще' больше не доступна.")
        break




```

```python
strategy_title = soup.find_all('a', class_ = "Z0rh")
for line in strategy_title: # Добавляем информацию в массив данных
    data['strategy_title'].append(line)
    href = line.get('href')
    data['href'].append(href)
print(data)
```
